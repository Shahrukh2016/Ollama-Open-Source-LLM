```python
## For chat conversions
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model= "ministral-3:3b",
    temperature= 0
)

response = llm.invoke("Why south pole is icy")
print(response)
```

```python
## For Non chat conversions
from langchain_ollama import OllamaLLM

llm = OllamaLLM(
    model= "ministral-3:3b"
)

response = llm.invoke("Capital of India is?", stream= True)

for i in response:
    print(i["content"], end= "")
```
