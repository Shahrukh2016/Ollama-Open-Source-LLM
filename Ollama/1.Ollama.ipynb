```python
 # GIVING MULTIMODEL INPUT TO MODEL [PASSING IMAGE TO THE MODEL]

import base64
import ollama

image_path= "Linkedin_Trend.jpg"

with open(image_path,"rb") as f: # OPENING THE IMAGE IN READ BINARY 
    image_bytes= f.read()    # READING THE IMAGE 
image_64= base64.b64encode(image_bytes).decode("utf-8") # ENCODING THE IMAGE IN BASE 64 BITS.

response= ollama.generate(model="ministral-3:3b", images=[image_64], prompt="Give caption for the image.")
print(response.response)
```

```python
# PASSING MULTIPLE IMAGES TO THE MODEL.

image_paths = ["Linkedin_Trend.jpg","Green AI.png"]

images_base64=[]
for i in image_paths:
    with open(i , "rb") as f:
        image_bytes= f.read()
        images_base64.append(base64.b64encode(image_bytes).decode("utf-8"))

response= ollama.generate(model="gemma3:4b", images=images_base64,
                          prompt="Generate an story based on these images, make sure you take context from each and every image.")
print(response.response)
```

```python
# SYSTEM INSTRUCTIONS
response = ollama.generate(model="ministral-3:3b",prompt="why does moon glow ?", system="You are an funny assistant , you explain things in funny way")
print(response.response)
```

```python
# OPTIONS PARAMETER
response = ollama.generate(model="ministral-3:3b",prompt="why is the ocean blue",
                           options={
                               "temperature":0.3,
                               "top_p":0.5,
                               "top_k":45
                           })
print(response.response)
```
